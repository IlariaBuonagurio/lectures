## Hands-on #5: Array creation

```python
a = np.ones((4, 4), dtype=int)
a[3,1] = 6
a[2,3] = 2

b = np.zeros((6, 5))
b[1:] = np.diag(np.arange(2, 7))

print(a)
print(b)
```

## Hands-on #6: Tiling for array creation

```python
block = np.array([[4, 3], [2, 1]])
a = np.tile(block, (2, 3))

print(a)
```

## Hands-on #7: reverse and reshape

```python
a = np.arange(10)

b = np.arange(5)

a[5:] = b[::-1]

print(a)
```

```python
a = np.arange(1, 16).reshape(3, -1).T
print(a)
```

## Hands-on #8: Prime number sieve

```python
is_prime = np.ones((100,), dtype=bool)
is_prime[:2] = 0
N_max = int(np.sqrt(len(is_prime)))
for j in range(2, N_max):
    if is_prime[j]: 
        is_prime[j*j::j] = False
print(np.nonzero(is_prime)[0])
```
